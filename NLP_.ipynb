# Stemming

The process of reducing words to their base or root form (not always a valid word). Example: "running", "runs"  "run"

```python
words=['running','runs','ran','easily','fairly','caring','studies','relational','happiness','organization']
```

## Porter Stemming

A classic stemming algorithm that reduces words to their root form by applying a series of rule-based suffix-stripping steps (e.g., "running"  "run").

```python
from nltk.stem import PorterStemmer
stemming=PorterStemmer()

for word in words:
  print(word," : ",stemming.stem(word))
```

## Regex Stemming

A customizable stemmer that uses regular expressions (regex) patterns to strip or replace word endings based on user-defined rules.

```python
from nltk.stem import RegexpStemmer
stemming=RegexpStemmer('ing$|s$|e$|able$|ly$|al$')

for word in words:
  print(word," : ",stemming.stem(word))
```

## Snowball Stemming

An improved and more flexible version of the Porter stemmer, written in the Snowball language, supporting multiple languages and offering more consistent results.

```python
from nltk.stem import SnowballStemmer
stemming=SnowballStemmer(language='english')

for word in words:
  print(word," : ",stemming.stem(word))
```

# Lemmatization

The process of reducing words to their base or dictionary form (lemma), using vocabulary and grammar rules. Example: "better"  "good", "running"  "run".

```python
from nltk import WordNetLemmatizer
nltk.download('wordnet')
```

```python
lemmatizer=WordNetLemmatizer()
for word in words:
  print(word," : ",lemmatizer.lemmatize(word, pos='v'))
```

```python
# pos = 'n' -> noun
# pos = 'v' -> verb
# pos = 'a' -> adjective
# pos = 'r' -> adverb
```

# Stopwords

Stopwords are the common words in a language (like the, is, and, of) that usually carry little meaningful information and are often removed from text during preprocessing to focus on the more important words.

```python
paragraph="""The quick brown fox jumps over the lazy dog. It is one of the most famous sentences in English because it contains every letter of the alphabet. However, in natural language processing, many words such as 'the', 'is', and 'of' are considered stopwords and are often removed."""
```

```python
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
```

```python
import nltk
```

```python
nltk.download('stopwords')
nltk.download('punkt_tab')
```

```python
stopwords.words('english')
```

```python
stemmer=PorterStemmer()
sentences=nltk.sent_tokenize(paragraph)
```

```python
sentences
```

```python
from nltk.tokenize import word_tokenize
for i in range(len(sentences)):
  words=word_tokenize(sentences[i])
  words=[lemmatizer.lemmatize(word, pos='v') for word in words if word not in set(stopwords.words('english'))]
  sentences[i]=' '.join(words)

print(sentences)
```

# One Hot Encoding

## Advantages

1. Easy to implement with python (`sklearn.OneHotEncoder()`, `pd.get_dummies()`)

## Disadvantages

1. Spase Matrix -> Overfitting
2. ML Algorithm -> Fixed Size Input
3. No Semantic Meaning is getting captured
4. Out of Vocabulary words (OOV)

## Code

```python
from sklearn.preprocessing import OneHotEncoder

# Corpus
corpus = [
    "the cat sat",
    "the dog barked",
    "the cat meowed"
]

# Tokenize
words = [[w] for sentence in corpus for w in sentence.split()]

# One-hot encode (dense output)
encoder = OneHotEncoder(sparse_output=False)
onehot = encoder.fit_transform(words)

print("Vocabulary:", encoder.categories_)
print("One-hot vectors:\n", onehot)
```

```python
all_words = [[w] for sentence in corpus for w in sentence.split()]

encoder.fit(all_words)

# Encode sentence by sentence
encoded_corpus = [
    encoder.transform([[w] for w in sentence.split()])
    for sentence in corpus
]

# Show results
for i, sent_matrix in enumerate(encoded_corpus):
    print(f"\nSentence {i+1}: {corpus[i]}")
    print(sent_matrix)
```
